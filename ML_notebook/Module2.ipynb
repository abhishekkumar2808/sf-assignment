## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

data=pd.read_csv('CinemaData.csv')
ax=plt.gca()
data.plot(kind='scatter',x='WeeksSinceRelease',y='OccPer',ax=ax)
plt.figure()
ax=plt.gca()
data.plot(kind='scatter',x='OtherReleasesInWeek',y='OccPer',ax=ax)
plt.figure()
ax=plt.gca()
data.plot(kind='scatter',x='OtherReleasesInWeek',y='Lifetime',ax=ax)
plt.figure()
ax=plt.gca()
data.plot(kind='scatter',x='ShowsInWeek',y='OccPer',ax=ax)
plt.figure()
ax=plt.gca()
data.plot(kind='scatter',x='ShowsInWeek',y='Lifetime',ax=ax)
```
